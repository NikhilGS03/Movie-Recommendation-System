```python
import pandas as pd
import numpy as np
```

```python
# read the csv file MovieGenre.csv
df = pd.read_csv('MovieGenre.csv', encoding = 'latin1')
df.head()
```

```python
# create A dataframe with the columns 'Title' and 'Genre' and IMDB score

df1 = df[['Title', 'Genre', 'IMDB Score']].head(1000)
df1.head(6)
```

```python
# split the Genre column by its delimiter '|'
df1['Genre'] = df1['Genre'].str.split('|')
df1.head()
```

```python
print(df1.tail())
```

```python
# dropping the na rows
# df = df[df['Genre'].notna()]
df1 = df1.dropna()

# separating the genres into different columns

# #For every row in the dataframe, iterate through the list of genres(which is in an array) and place a 1 into the corresponding column 

for index, row in df1.iterrows():
    # print(row)
    for genre in row['Genre']:
        # print(type(row['Genre']))   
        df1.at[index, genre] = 1

# fill in the NaN values with 0 to show that a movie doesn't have that column's genre
df1 = df1.fillna(0)
df1.head()
```

```python
# create user input
# movies user 1 has watched
userInput = [
            {'Title':df1['Title'].loc[df1.index[0]], 'Rating':6},
            {'Title':df1['Title'].loc[df1.index[1]],  'Rating':8},
            {'Title':df1['Title'].loc[df1.index[2]], 'Rating':2},
            {'Title':df1['Title'].loc[df1.index[7]],  'Rating':9},
            {'Title':df1['Title'].loc[df1.index[9]],  'Rating':5},
            {'Title':df1['Title'].loc[df1.index[10]],  'Rating':4}
         ] 

# create a dataframe with the movies and the rating
inputMovies = pd.DataFrame(userInput)
inputMovies

# u1_ratings = np.array([inputMovies['Rating'].to_numpy()])
u1_ratings = inputMovies['Rating'].to_numpy()

print(u1_ratings,u1_ratings.shape)
```

```python
# normalizing imdb score
df1['IMDB Score'] = df1['IMDB Score'].map(lambda score: score/10)
```

```python
# final movie genre matrix

f_df = df1.drop(['Title','Genre'],axis='columns')
f_df.head()
```

```python
# getting the genre matrix for the movies user1 has watched
# mask = np.zeros(f_df.shape[0])
idx = [0,1,2,7,9,10]
# for id in idx:
#     mask[id]=1

# print(mask)
C1 = f_df.iloc[idx]
# # .to_numpy()
print(C1)
print(C1.shape)
```

```python
# to get user 1's genre weighted matrix

print(f"the dimensions are {u1_ratings.shape} and {C1.shape}")

u1_weights = np.matmul(u1_ratings, C1)
# print(u1_weights)

sum=0
for i in u1_weights:
  sum+=i
# print(u0_weights,sum)
final_genre_vector=[]
for i in range(len(u1_weights)):
  # print(i)
  # print(u0_weights[0][i]/sum)
  final_genre_vector.append(u1_weights[i]/sum)
final_genre_vector = np.array(final_genre_vector)
print("Your final weighted genre matrix for user1 vector is:",final_genre_vector, final_genre_vector.shape,sep="\n")
```

```python
# for the new movies

p_df = f_df.loc[11:61]
p_df.head()

```

```python
C2 = p_df.to_numpy()
print(C2,C2.shape,sep="\n")
```

```python
# so now we have to multiply our user genre weighted matrix which is in final_genre_vector and our new movie matrix which is in C2

print("the dimensions are: \n", final_genre_vector.shape, C2.shape)

C2t = C2.transpose()
print(C2t.shape)
```

```python
# so our new multiplication will now be final_genre_vector * C2t

aans = np.matmul(final_genre_vector,C2t)
print(aans,'\n',aans.shape)
```

```python
aans = (aans.argsort())[50:30:-1]
print(aans)

pred_arr = []

for i in aans:
    pred_arr.append(df1.Title[i+11])

print(pred_arr)
```
